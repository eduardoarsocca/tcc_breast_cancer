# 5. Tratamento dos dados

```python
brest_cancer = brest_cancer.drop(columns=['id', 'Unnamed: 32'])
brest_cancer.head(2)
```

```python
brest_cancer.dtypes
```

Como é possível verificar, todos as variáveis são referentes à análise moforlógica dos componentes celulares.

As variáveis em questão descreevm as características dos núcleos presentes nas imagens obtidas do aspirado de células.

variáveis:
Diagnóstico -> M = Maligno, B = Benigno.
raio -> (média das distâncias do centro aos pontos do perímetro nuclear).
textura -> aspecto morfológico do núcleo (descio padrão dos valores da escala de cinza).
perímetro -> tamanho médio do tumor central.
área
suavidade -> média da variação local dos comprimentos de raio.
compacidade -> severidade dos seguimentos concavos do contorno nuclear.
pontos côncavos -> número de seguimentos côncavos do contorno nuclear.
simetria
dimensão fractal -> medida da complexidade estrutural das células tumorais.

```python
brest_cancer_limpo = brest_cancer[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
matriz_correlacao = brest_cancer_limpo.corr()
plt.figure(figsize=(16,12))
sns.heatmap(matriz_correlacao, annot=True, fmt = '.2f', cmap = 'coolwarm')
plt.title('Matriz de correlação entre os parâmetros')
plt.show()
```

```python
cancer_maligno = brest_cancer[brest_cancer.diagnosis ==  'M']
cancer_benigno = brest_cancer[brest_cancer.diagnosis ==  'B']

sns.set(style='whitegrid')

parametros_selecinados = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst'
       ]

fig, axes = plt.subplots(len(parametros_selecinados), 1, figsize = (10,45))

for i, parametros in enumerate(parametros_selecinados):
    sns.histplot(cancer_maligno[parametros], color = 'red', label = 'Maligno', ax = axes[i], kde = True, stat = 'density', linewidth = 0)
    sns.histplot(cancer_benigno[parametros], color = 'blue', label = 'Benigno', ax = axes[i], kde = True, stat = 'density', linewidth = 0)
    axes[i].set_title(f'Distribuição de {parametros}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

Avaliando a matriz de correlação, bem como os gráficos de distribuição para cada um dos parâmetros da base, é possivel verificar que alguns dos parâmetros mostram diferenças notáveis entre as classes de câncer maligno (em vermelho) e benigno (em azul). Os parâmetros em que não há sobreposição estatisticamente significativa entre os resultados para os dois tipos de canceres podema judar na melhor categorização dos tipos de células neoplásicas e seu comportamento.
